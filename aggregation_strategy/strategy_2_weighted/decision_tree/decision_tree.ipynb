```python
import joblib

from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import export_text
```

```python
tree_model = joblib.load("decision_tree_model.joblib")
```

```python
print(tree_model.get_params())        
print(tree_model.feature_importances_) 
print(tree_model.classes_)           
print(tree_model.tree_.max_depth)     
print(tree_model.tree_.node_count)  
```

```python
plt.figure(figsize=(12, 8))
tree.plot_tree(tree_model, feature_names=["feat1", "feat2", "feat3"], class_names=True, filled=True)
plt.show()
```
